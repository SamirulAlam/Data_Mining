```python
def plot_from_dictionary(dic, title=None, size=(13.5, 8.5), bottom=0.1, x_font='kalpurush', save_to_device=True):
    fig = plt.figure(figsize=size)  # Size of the chart windows
    fig.subplots_adjust(bottom=bottom)

    barChart = plt.bar(list(dic.keys()), dic.values())  # Bar chart
    plt.title(title, fontsize=28)  # Chart Title
    plt.style.use('ggplot')



    # X-axes & Y-axes labels
    plt.xticks(list(dic.keys()), dic.keys(), fontname=x_font, rotation='vertical')
    plt.yticks(np.arange(0, max(dic.values()) + max(dic.values()) // 4, max(dic.values()) // 10))

    for i in range(len(list(dic.keys()))):
        plt.text(i, list(dic.values())[i] + max(0.5, max(dic.values()) // 40), list(dic.values())[i], ha='center',
                 rotation='vertical')  # Adding text on each bar columns
    plt.show()  # Display plot





    if save_to_device:
        fig.savefig('Uni_gram')
```

```python
data = pd.read_csv("ShamsurRahmanKobita.csv")
all_data = []
unique = []

df = data["Poem"].str.split(" ")
```

```python
for word in df:
    for x in word:
        result = re.sub("[:+—–_(){}\n;|.,?’!@<>।॥“”'\"\\[\\]]", ' ', x)
        all_data.append(result)
```

```python
for char in all_data:
    if char not in unique:
        unique.append(char)
```

```python
wordDict = {}
for word in all_data:
    try:
        wordDict[word] += 1
    except KeyError:
        wordDict[word] = 1

wordDict = dict(sorted(wordDict.items(), key=lambda item: item[1]))
resDict = {}
```

```python
for word in wordDict:
    if  wordDict[word] >= 1:
        resDict[word] = wordDict[word]
```

```python
samir = {'letter': resDict.keys(), 'frequency': resDict.values() }




df = pd.DataFrame(samir)
df.to_csv('unigram.csv')
```

```python
print(wordDict)
plot_from_dictionary(dict(zip(list(resDict.keys())[-50:], list(resDict.values())[-50:])),
                         title='Unigram of Dataset',
                         save_to_device=True)
```
