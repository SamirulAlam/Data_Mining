```python
import os
import pandas as pd

# Folder containing the CSV files
folder_path = 'D:\Python_Projects\Dm\Dataset'  # Replace with the actual path

# List to store dataframes
dataframes = []

# Loop through the CSV files in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):
        file_path = os.path.join(folder_path, file_name)
        df = pd.read_csv(file_path, on_bad_lines='skip')
        dataframes.append(df)

# Assign dataframes to variables
for i, df in enumerate(dataframes):
    var_name = f'df{i+1}'  # Naming variables as df1, df2, ...
    globals()[var_name] = df

# Now you have df1, df2, ..., dfN as variables containing the data from the CSV files
```

```python
import random

# List of variable names
var_names = ['df1', 'df2', 'df3', 'df4', 'df5', 'df6', 'df7', 'df8', 'df9', 'df10']

# Randomly select 5 variable names
selected_var_names = random.sample(var_names, 5)

# Access the selected dataframes using the selected variable names
selected_dataframes = [globals()[var_name] for var_name in selected_var_names]

# Display the head of each selected dataframe
for i, df in enumerate(selected_dataframes):
    print(f"Head of {selected_var_names[i]}:")
    print(df.head())
    print('-' * 30)
```
